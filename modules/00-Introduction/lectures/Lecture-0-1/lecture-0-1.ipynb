```python
import pandas as pd
import numpy as np
from lets_plot import *
LetsPlot.setup_html()

import vega_datasets
cars = vega_datasets.data.cars()
```

```python
nums = (np.random.random(10) * 100).astype(int)
cats = ['a'] * 5 + ['b'] * 5
ind = list(np.arange(5)) + list(np.arange(5)) 
data = pd.DataFrame(dict(count=nums, cat=cats, ind=ind))

ggplot() + geom_bar(aes(x='ind', y='count'), data=data, \
        stat='identity', show_legend=False, fill='white', color='black', stroke=2) + \
        scale_fill_discrete() + facet_grid(x='cat') + ggsize(400, 400)
```
